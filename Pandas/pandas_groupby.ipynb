```python
# Find no. of  movie starts with A in every genre
```

```python
def foo(group):
    return(((group['Series_Title'].str.startswith('A')) | (group['Series_Title'].str.startswith('a'))).sum())
```

```python
genre.apply(foo)
```

```python
# Finding Ranking of each movie in the group to IMDB score

def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genre.apply(rank_movie)
```

```python
# Find Normalize Rating group wise

def norm(group):
    group['norm_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genre.apply(norm)
```

```python
# Create a Function that can return the highest score of any batsman
ipl.columns
```

```python
def max_run(batsman):
    return ipl[ipl['batsman'] == batsman].groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(1).values[0][1]
```

```python
ipl.groupby(['batsman','inning'])['batsman_runs'].max().sort_values(ascending=False)
```

```python
max_run('CA Lynn')
```
